```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('yearincome.csv')
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year, df.income, color='r', marker='.')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year, df.income, color='r', marker='.')
plt.plot(df.year,reg.predict(df[['year']]))
```
